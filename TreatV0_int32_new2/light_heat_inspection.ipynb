```c++
Long64_t HeatPos, LightPos;
Double_t rawamp, light_rawamp;  // 定义 rawamp 和 light_rawamp
Int_t IsLight;  // 定义 IsLight 变量

// 设置分支的地址
tree->SetBranchAddress("trigpos", &HeatPos);
tree->SetBranchAddress("light_trigpos", &LightPos);
tree->SetBranchAddress("rawamp", &rawamp);
tree->SetBranchAddress("light_rawamp", &light_rawamp);
tree->SetBranchAddress("IsLight", &IsLight);  // 关联 IsLight 分支
```

```c++
// 创建第一个画布和 LightDelay 的一维直方图
TCanvas *canvas1 = new TCanvas("canvas1", "light_delay", 800, 600);
TH1F *hLightDelay = new TH1F("hLightDelay", "Light Signal Delay", 150, -50, 100);

// 创建第二个画布和 RawAmp vs Light_RawAmp 的二维直方图
TCanvas *canvas2 = new TCanvas("canvas2", "RawAmp vs Light_RawAmp", 800, 600);
TH2F *h2DAmp = new TH2F("h2DAmp", "RawAmp vs Light_RawAmp (30 < LightDelay < 33)", 
                        350, 0, 35000, 100, 0, 2000);  // 二维直方图：rawamp 范围 [0, 20000]，light_rawamp 范围 [0, 2000]

// 遍历TTree并填充直方图
Long64_t nentries = tree->GetEntries();
for (Long64_t i = 0; i < nentries; ++i) {
    tree->GetEntry(i);

    // 计算 LightDelay
    Double_t LightDelay = LightPos - HeatPos;
    // 仅当 IsLight == 1 且 LightDelay 在 30 到 33 之间时填充二维直方图
    if (IsLight == 1) {
    // 填充 LightDelay 的一维直方图
    hLightDelay->Fill(LightDelay);
        // if (LightDelay >= 30 && LightDelay <= 33) {
        if (LightDelay >= 20 && LightDelay <= 40) {
        h2DAmp->Fill(rawamp, light_rawamp);
    }
    }
}

// 设置 LightDelay 的一维直方图标题和轴标签
hLightDelay->SetTitle("Light Signal Delay");
hLightDelay->GetXaxis()->SetTitle("Points");
hLightDelay->GetYaxis()->SetTitle("Counts/points");

// 绘制 LightDelay 的一维直方图
canvas1->cd();
hLightDelay->Draw();

// 设置 RawAmp vs Light_RawAmp 的二维直方图标题和轴标签
// h2DAmp->SetTitle("RawAmp vs Light_RawAmp (30 < LightDelay < 33)");
h2DAmp->SetTitle("RawAmp vs Light_RawAmp (20 < LightDelay < 40)");
h2DAmp->GetXaxis()->SetTitle("RawAmp");
h2DAmp->GetYaxis()->SetTitle("Light_RawAmp");
h2DAmp->SetMarkerStyle(20);
h2DAmp->SetMarkerSize(0.8);

// 绘制 RawAmp vs Light_RawAmp 的二维直方图
canvas2->cd();
h2DAmp->Draw("COLZ");  // 使用 "COLZ" 绘制二维直方图
```
